# 1. Sigmoid

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 100)  
sig_x = sigmoid(x)  
print(x)
print(sig_x)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.plot(x, sig_x, label='Sigmoid Function', color='blue')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```

# 2. Tanh (Hyperbolic Tangent)

```python
def tanh(x):
    return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))

tanh_y = tanh(x)
print(tanh_y)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x, tanh_y, label='Tanh Function', color='red')
plt.title('Tanh Function')
plt.xlabel('x')
plt.ylabel('Tanh(x)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```
